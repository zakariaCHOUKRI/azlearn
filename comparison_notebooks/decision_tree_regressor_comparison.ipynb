```python
import sys
sys.path.insert(1, '../azlearn')

from tree_model import DecisionTreeRegressor as TreeRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

# Load the diabetes dataset
data = load_diabetes()
X = data.data
y = data.target

# Train decision tree model
tree_model = TreeRegressor.DecisionTreeRegressor()
tree_model.fit(X, y)

dtree_model = DecisionTreeRegressor(random_state=42)
dtree_model.fit(X, y)

# Make predictions
y_pred = tree_model.predict(X)
y_pred2 = dtree_model.predict(X)

# Print predictions
print("Predictions comparison (first 20 samples)")
for i in range(min(len(y), 20)):
    print(f"Actual: {y[i]}, Our model: {y_pred[i]}, scikit-learn model: {y_pred2[i]}")

# Plot actual vs predicted values for your model
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred, color='blue', label='Our Model')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted values (Custom Decision Tree Regressor)')
plt.legend()
plt.show()

# Plot actual vs predicted values for scikit-learn model
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred2, color='green', label='Scikit-learn Model')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted values (Scikit-learn Decision Tree Regressor)')
plt.legend()
plt.show()
```
