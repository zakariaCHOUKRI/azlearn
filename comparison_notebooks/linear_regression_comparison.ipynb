```python
import sys
sys.path.insert(1, '../azlearn')

from linear_model.LinearRegression import LinearRegression as MyLinearRegression
from sklearn.linear_model import LinearRegression as SklearnLinearRegression
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

# Load the diabetes dataset
data = load_diabetes()
X = data.data  # Using all features
y = data.target

print("comparing linear regression models")

myModel = MyLinearRegression()
sklModel = SklearnLinearRegression()

myModel.fit(X, y)
sklModel.fit(X, y)

y_pred = myModel.predict(X)
y_pred2 = sklModel.predict(X)

# Plot actual vs predicted values for your model
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred, color='blue', label='Our Model')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted values (Custom Linear Regression)')
plt.legend()
plt.show()

# Plot actual vs predicted values for scikit-learn model
plt.figure(figsize=(10, 6))
plt.scatter(y, y_pred2, color='green', label='Scikit-learn Model')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted values (Scikit-learn Linear Regression)')
plt.legend()
plt.show()
```
