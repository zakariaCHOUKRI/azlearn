```python
import sys
sys.path.insert(1, '../azlearn')

from neural_network.MLP import MLPRegressor as MyMLPRegressor
from sklearn.neural_network import MLPRegressor as SklearnMLPRegressor
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(20, 1)  # 100 samples, 1 feature
y = 2 * X.squeeze() + 3 + np.random.randn(20)  # true relationship: y = 2*X + 3 + noise

print("comparing linear regression models")

myModel = MyMLPRegressor()
sklModel = SklearnMLPRegressor()

myModel.fit(X, y.reshape(-1, 1))  # Reshape y to have two dimensions
sklModel.fit(X, y)

y_pred = myModel.predict(X)
y_pred2 = sklModel.predict(X)

# Plot actual vs predicted values for your model
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Our Model Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Actual vs Predicted values (Custom MLP Regressor)')
plt.legend()
plt.show()

# Plot actual vs predicted values for scikit-learn model
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred2, color='green', label='Scikit-learn Model Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Actual vs Predicted values (Scikit-learn MLP Regressor)')
plt.legend()
plt.show()
```
